```python
#3
import math

def normal_pdf(x, mean, std_dev):

    coefficient = 1 / (math.sqrt(2 * math.pi) * std_dev)
    exponent = -(1/2) * ((x - mean) / std_dev)**2
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

mean = 0
std_dev = 1
x_value = 1.5

pdf_result = normal_pdf(x_value, mean, std_dev)
print(f"The PDF at x={x_value} is: {pdf_result}")
```

```python
#6
import math

def poisson_cdf(x, mean):

    cdf_value = 0
    for k in range(x + 1):
        cdf_value += (math.exp(-mean) * mean**k) / math.factorial(k)
    return cdf_value


mean_poisson = 3
x_value_poisson = 5

cdf_result_poisson = poisson_cdf(x_value_poisson, mean_poisson)
print(f"The CDF at x={x_value_poisson} is: {cdf_result_poisson}")
```
