```python
#Q.3
"""The subplot function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or axes in a single figure,
 making it easier to compare different visualizations or aspects of your data. The subplot function takes three integer arguments: the number of rows, the number of columns,
  and the index ."""
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

y2 = np.array([50, 20, 40, 20, 60, 70])

y3 = np.array([10, 20, 30, 40, 50, 60])

y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Subplot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Subplot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Subplot 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Subplot 4')
plt.legend()



plt.tight_layout() #it is use for better spacing


plt.show()

```

```python
#Q.4
"""A bar plot is a graphical representation of data where rectangular bars of varying
  lengths are used to represent different categories or groups.Bar plots are commonly used for visualizing categorical data
  and comparing values across different categorie"""
"""use of bar plot
    1.comparision of categories
    2.categorial Distribution
    3.trend Analysis
"""
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("company name")
plt.ylabel("profits")
plt.bar(company,profit)


```

```python
#Q.5
""" A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution
    of a dataset. It provides a visual summary of key statistics such as the minimum, first quartile (25th percentile),
    median (50th percentile), third quartile (75th percentile), and maximum. The box plot is composed of several components."""
 #uses of box plot
"""
      1.he lines extending from the box (sometimes called whiskers) represent the minimum and maximum values within the dataset .
      2.he line inside the box represents the median, which is the middle value when the dataset is sorted.
"""
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
data_to_plot=[box1,box2]
plt.boxplot(data_to_plot,labels=['Box1','Box2'])




```
